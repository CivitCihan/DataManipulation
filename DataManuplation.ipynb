<h1>Data Analysis and Virtualization with Pandas and Seaborn</h1>

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = sns.load_dataset("planets")
```

```python
df.head()
```

```python
df.shape
```

```python
df.plot(figsize=(12,6), x = "mass", y = "distance", kind = "scatter")
plt.title("Planets")
plt.xlabel("Mass")
plt.ylabel("Distance")
plt.show
```

```python
df.groupby("method").mean()
```

```python
df.groupby("method")["distance"].mean()
```

```python
df.groupby("method").describe()
```

```python
df.groupby("method").aggregate([np.median, "min", "max"])
```

```python
df.groupby("method").aggregate({"mass": "max", "distance": "min", "year": "mean" })
```

```python
def filter_func(x):
    return x["distance"].std() > 1000
```

```python
df.groupby("method").filter(filter_func)
```
